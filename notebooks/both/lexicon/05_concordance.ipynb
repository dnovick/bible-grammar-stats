# Concordance: Finding All Occurrences of a Lemma

The concordance module searches for every occurrence of a Hebrew or Greek lemma across the OT (TAHOT), NT (TAGNT), or LXX. Results include the verse reference, the inflected word form, morphological details, and an English verse for context.

**Three corpora are available:**
- `corpus='OT'`  — Hebrew/Aramaic OT (TAHOT, ~425k words)
- `corpus='NT'`  — Greek NT (TAGNT, ~138k words)  
- `corpus='LXX'` — Greek Septuagint (CenterBLC/LXX, ~623k words)

````python
import sys
sys.path.insert(0, '../../../src')

import pandas as pd
pd.set_option('display.max_colwidth', 120)
pd.set_option('display.max_rows', 60)

from bible_grammar.concordance import concordance, lemma_frequency, top_lemmas
````

## 1. Hebrew Concordance: בָּרָא (H1254, "to create")

````python
df = concordance(strongs='H1254', context='KJV')
print(f"Total occurrences: {len(df)}")
df[['reference', 'word', 'stem', 'conjugation', 'context_text']]
````

### Distribution by book

````python
lemma_frequency(strongs='H1254')
````

### Niphal occurrences only (passive "to be created")

````python
df_nip = concordance(strongs='H1254', stem='Niphal', context='KJV')
print(f"Niphal occurrences: {len(df_nip)}")
df_nip[['reference', 'word', 'stem', 'conjugation', 'context_text']]
````

## 2. Hebrew Concordance with Hebrew Text Context

Use `context='Hebrew'` to see the full Hebrew verse instead of KJV.  
Example: נָגַד (H5046) in the Hiphil stem ("to declare").

````python
# נָגַד (H5046) Hiphil — "to declare / tell"
df_ngd = concordance(strongs='H5046', stem='Hiphil', context='Hebrew', book_group='torah')
print(f"Hiphil of H5046 in Torah: {len(df_ngd)}")
df_ngd[['reference', 'word', 'stem', 'conjugation', 'context_text']].head(10)
````

## 3. Greek NT Concordance: ποιέω (G4160, "to do / make")

````python
df_nt = concordance(strongs='G4160', corpus='NT', context='KJV')
print(f"Total NT occurrences: {len(df_nt)}")
df_nt[['reference', 'word', 'tense', 'voice', 'mood', 'context_text']].head(15)
````

### Frequency by NT book

````python
lemma_frequency(strongs='G4160', corpus='NT')
````

### Aorist Active Indicative forms only

````python
df_aor = concordance(strongs='G4160', corpus='NT', context='KJV',
                     book_group='gospels')
df_aor_ind = df_aor[(df_aor['tense']=='Aorist') & 
                    (df_aor['voice']=='Active') & 
                    (df_aor['mood']=='Indicative')]
print(f"Aorist Active Indicative of ποιέω in Gospels: {len(df_aor_ind)}")
df_aor_ind[['reference', 'word', 'context_text']].head(10)
````

## 4. LXX Concordance: ποιέω in Genesis

````python
df_lxx = concordance(strongs='G4160', corpus='LXX', book='Gen', context='KJV')
print(f"ποιέω in LXX Genesis: {len(df_lxx)}")
df_lxx[['reference', 'word', 'lemma', 'tense', 'voice', 'context_text']].head(15)
````

## 5. Top Lemmas in a Corpus

`top_lemmas()` shows the most frequent Strong's numbers, useful for frequency analysis.

````python
# Top OT Hebrew verbs
print("Top 15 OT Hebrew verbs (by occurrence):")
top_lemmas(corpus='OT', part_of_speech='Verb', top_n=15)
````

````python
# Top NT Greek verbs in Paul's letters
print("Top 15 Pauline Greek verbs:")
top_lemmas(corpus='NT', part_of_speech='Verb', book_group='pauline', top_n=15)
````

## 6. Quick Reference

```python
from bible_grammar.concordance import concordance, lemma_frequency, top_lemmas

# All occurrences of a Hebrew root
concordance(strongs='H1254')                          # KJV context by default
concordance(strongs='H1254', context='Hebrew')        # show Hebrew verse text
concordance(strongs='H1254', stem='Niphal')           # filter by binyan
concordance(strongs='H1254', book_group='prophets')   # restrict to Major/Minor Prophets

# NT Greek
concordance(strongs='G4160', corpus='NT')
concordance(strongs='G2316', corpus='NT', book='Rom') # θεός in Romans

# LXX Greek
concordance(lemma='κύριος', corpus='LXX')             # by Unicode lemma
concordance(lemma_translit='kurios', corpus='LXX')    # by transliteration
concordance(strongs='G2962', corpus='LXX')            # by Strong's

# Frequency distribution
lemma_frequency(strongs='H5046')                      # counts per OT book
lemma_frequency(strongs='G4160', corpus='NT')         # counts per NT book

# Most frequent lemmas
top_lemmas(corpus='OT', part_of_speech='Verb', book_group='torah', top_n=20)
top_lemmas(corpus='NT', part_of_speech='Noun', book_group='pauline', top_n=20)
```
