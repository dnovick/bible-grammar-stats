# Word Study Tool

The `word_study` module produces a complete one-stop profile for any Hebrew or Greek lemma. For each Strong's number it combines:

- **Lexicon entry** — lemma, transliteration, gloss, and full definition from STEPBible's TBESH (Hebrew) or TBESG (Greek) lexicons
- **Total occurrences** and **distribution by book**
- **Morphological form inventory** — all stem/conjugation combinations used (Hebrew) or tense/voice/mood (Greek)
- **LXX translation equivalents** — which Greek lemmas the LXX uses to render this Hebrew word (OT only)
- **Example verses** with KJV English context

## 5. Programmatic access: word_study() returns a dict

````python
ws = word_study('H1254')

print(f"Lemma: {ws['lemma']}  ({ws['translit']})")
print(f"Gloss: {ws['gloss']}")
print(f"Total occurrences: {ws['total_occurrences']}")
print()
print("Distribution by book:")
display(ws['by_book'])
print("\nMorphological forms:")
display(ws['morphological_forms'])
print("\nLXX translation equivalents (top 10 — filtered to Verbs):")
from bible_grammar.alignment import translation_equivalents
display(translation_equivalents(heb_strongs='H1254', lxx_pos='Verb', top_n=10))
````

## 6. Tabular concordance form

`word_study_table()` returns every occurrence as a DataFrame row — equivalent to `concordance()` but with the lexicon gloss attached.

````python
tbl = word_study_table('H1254')
print(f"Lemma: {tbl.attrs.get('lemma', '')}  —  {tbl.attrs.get('gloss', '')}")
tbl[['reference', 'word', 'stem', 'conjugation', 'context_text']]
````

## Quick Reference

```python
from bible_grammar.wordstudy import word_study, print_word_study, word_study_table

# Formatted output to console
print_word_study('H1254')    # Hebrew: בָּרָא "to create"
print_word_study('H3519')    # Hebrew: כָּבוֹד "glory"
print_word_study('H8451')    # Hebrew: תּוֹרָה "law, instruction"
print_word_study('G3056')    # Greek: λόγος "word"
print_word_study('G4102')    # Greek: πίστις "faith"
print_word_study('G26')      # Greek: ἀγάπη "love"
print_word_study('G5547')    # Greek: Χριστός "Christ"

# Get dict for programmatic access
ws = word_study('H8451')
ws['lemma']                  # תּוֹרָה
ws['gloss']                  # "law, instruction"
ws['definition']             # full BDB/LSJ definition
ws['total_occurrences']      # int
ws['by_book']                # DataFrame: book_id, book_name, count, pct
ws['morphological_forms']    # DataFrame: stem/conjugation or tense/voice/mood
ws['translation_equivalents'] # DataFrame: LXX Greek lemmas (Hebrew only)
ws['examples']               # list of {reference, word, context} dicts

# Full concordance with lexicon gloss attached
word_study_table('G4102')    # all πίστις occurrences
```
