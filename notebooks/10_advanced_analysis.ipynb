## 1. Divine Names Analysis

The `divine_names` module tracks the major divine names and christological titles
across the Hebrew OT and Greek NT:

Elohim is proportionally more prominent in the Wisdom literature; in the NT, Christos
and Iesous are nearly absent from Revelation's prophetic poetry compared with the Pauline
epistles. These shifts reflect genre conventions, theological emphasis, and the
communities for which each corpus was written.

## 2. Genre Comparison

The `genre_compare` module measures how morphological patterns shift across the
canonical literary sections of the Hebrew OT and Greek NT.

**OT genres:** Torah, Historical, Wisdom, Prophets  
**NT genres:** Gospels & Acts, Pauline, General & Rev

**Key finding:** The Consecutive Perfect (wayyiqtol) conjugation — the classical
Hebrew narrative waw-consecutive — accounts for roughly **36%** of all verbs in the
Historical books but drops to only **~5%** in Wisdom literature. This single statistic
confirms what grammarians have long observed: wayyiqtol is the backbone of Hebrew
prose narrative, while Wisdom poetry prefers the timeless Imperfect and Participle to
express ongoing states and generalizations rather than sequential events.

````python
from bible_grammar import print_genre_compare

# OT verb stem distribution: how Qal, Piel, Hiphil, Niphal, etc. vary by genre
print_genre_compare('OT', feature='verb_stem')
````

````python
# OT verb conjugation: note how wayyiqtol (Consecutive Perfect) spikes in Historical
print_genre_compare('OT', feature='verb_conjugation')
````

````python
# NT verb tense: Aorist dominates narrative; Present dominates epistolary writing
print_genre_compare('NT', feature='verb_tense')
````

## 3. Intertextuality Network

The `intertextuality` module maps OT passages → NT quotations and allusions, using
the scrollmapper cross-reference dataset (OpenBible.info, CC-BY).

**Vote scores** reflect community confidence: each cross-reference link in the dataset
is scored by how many independent annotators identified the connection. A score of 20
means at least 20 votes of confidence; 50+ represents very high consensus. Higher
scores generally track with direct verbal quotation, while lower scores (5–15) may
indicate a thematic allusion or echo.

**Bipartite layout:** In the network graph, **OT verses** appear as blue squares on the
left column and **NT books/verses** appear as coral circles on the right column. Directed
edges run from each OT verse to the NT location that quotes or alludes to it; edge
thickness and opacity scale with the vote score, so high-confidence links stand out
visually. This bipartite arrangement makes it easy to spot which OT verses are most
heavily cited and which NT books draw on the same OT passage.

````python
from bible_grammar import print_intertextuality

# Isaiah 53 — the Suffering Servant chapter; among the most-cited OT passages in the NT
print_intertextuality('Isa', chapter=53, min_votes=20)
````

````python
# Psalm 22 — the lament Psalm quoted by Jesus from the cross
print_intertextuality('Psa', chapter=22, min_votes=15)
````
